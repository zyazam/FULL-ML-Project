### Methodology: Ridge Regression (L2 Regularization)
L2 Regularization: This model adds a penalty term (squared magnitude of coefficients) to the loss function to prevent the model from over-relying on any single feature.

Multicollinearity Handling: Ridge is specifically used to address potential multicollinearity issues (high correlation between features), which often leads to unstable coefficients in standard OLS.

Complexity Control: By shrinking the coefficients towards zero, it helps in reducing model variance and improving the generalization performance on the test set.

```python
# --- 2. Ridge Regression (L2 Regularization) ---
# Shrinking coefficients to manage multicollinearity and prevent overfitting.
ridge = Ridge(alpha=1.0)
ridge.fit(X_train_numeric, y_train)
ridge_pred = ridge.predict(X_test_numeric)
evaluate_regression("Ridge (L2)", y_test, ridge_pred, X_test_numeric)
```
