### Methodology: Lasso Regression (L1 Regularization)
L1 Regularization: Unlike Ridge, Lasso adds a penalty proportional to the absolute value of the coefficients.

Automatic Feature Selection: A key characteristic of Lasso is its ability to shrink the coefficients of less important features exactly to zero, effectively performing feature selection.

Model Sparsity: This process creates a "sparse model" that focuses only on the most influential predictors, which can be highly beneficial for understanding which behavioral features truly drive the days_since_prior_order.

Interpretability: By eliminating redundant or noisy features, Lasso helps in simplifying the model without sacrificing significant predictive power.

```python
# --- 3. Lasso Regression (L1 Regularization) ---
# Using L1 penalty for automatic feature selection by zeroing out insignificant weights.
lasso = Lasso(alpha=0.1)
lasso.fit(X_train_numeric, y_train)
lasso_pred = lasso.predict(X_test_numeric)
evaluate_regression("Lasso (L1)", y_test, lasso_pred, X_test_numeric)
```
