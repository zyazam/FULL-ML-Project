   - ### ****The model was implemented using Logistic Regression with L2 Regularization (Ridge).    - ### ****Precision (0.1550): The low precision is a direct result of the balanced         weight strategy and the sparsity of the target class. The model intentionally             produces more False Positives to maximize the discovery of True Positives****

   - ### ****MCC (0.2502): The Matthews Correlation Coefficient confirms that the model’s    predictions are significantly better than random guessing, providing a reliable           measure for this imbalanced scenario****
